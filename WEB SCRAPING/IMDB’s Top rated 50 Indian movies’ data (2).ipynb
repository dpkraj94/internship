```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
All_movies = []

for i in soup.find_all('td',class_="titleColumn"):
    All_movies.append(i.text)
    
All_movies
```

```python
rating = []

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
    
rating
```

```python
release_year = []

for i in soup.find_all('span',class_="secondaryInfo"):
    release_year.append(i.text)
    
release_year 
```

```python
print(len(All_movies),len(rating),len(release_year))
```

```python
import pandas as pd
df = pd.DataFrame({'Movie name':All_movies[0:50],'Rating':rating[0:50],'Release year':release_year[0:50]})
df
```
